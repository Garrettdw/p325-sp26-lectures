# User Input and While Loops

## Practice

When buying tickets to a movie, your age determines which movies you're allowed to see (G, PG, PG-13, R). Write a `while` loop that asks a user for their age, then prints the highest rated movie they are allowed to see. Include a quit prompt that will use `break` to stop the loop. Can you think of a reasonable use for the `continue` statement in this example?

```python
Prompt = 'Please enter your age (enter) q to quit): '

active = True
while active:
    response = input(Prompt)
    if response == 'q':
        active = False
        break
    age = float(response)
    if age <= 8:
        print("you are permitted to see G rated movies")
    elif 8 < age < 13:
        print("you are permitted to see PG rated movies")
    elif 13<= age < 18:
        print('you are permitted to see PG-13 reted movies')
    else:
        print('you are premitted to see R rated movies')
```
